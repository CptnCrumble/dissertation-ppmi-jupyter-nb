```python
import pandas
import numpy
import matplotlib.pyplot as plot

data_dictionary = pandas.read_csv('source_data/Data_Dictionary_-_Harmonized.csv')

def definerer(itm_name):
    tmp = data_dictionary[data_dictionary['ITM_NAME'] == itm_name]
    if tmp.size > 0:
        print(itm_name)
        print(tmp.iloc[0]['DSCR'])
        print('---')
    else:
        print(itm_name + ' not in data dictionary\n---')

def df_deets(dframe):
    for x in dframe.columns:
        definerer(x)
        
def make_a_set(dframe,column):
    output = set()
    for x in dframe[column]:
        output.add(x)
    return output
```

```python
updrs_ii = pandas.read_csv('source_data/MDS_UPDRS_Part_II__Patient_Questionnaire.csv')
falls = pandas.read_csv('source_data/Determination_of_Freezing_and_Falls.csv')

updrs_sigfall_base = pandas.read_csv('working_data/updrs_sigfall_base.csv')
```

```python
# f(x) : PATNO,date -> fall within 12months?
# yes but WHAT IS a fall ??

# Create new binary flag SIGFALL (significant fall)
# what are the interesting columns?
# FLLDRVIS
# Fall resulted in outpatient visit to a healthcare provider
# ---
# FLLERVIS
# Fall resulted in ER visit
# ---
# FLLHOSP
# Fall resulted in hospitalization
# ---
# FLLSURG
# Fall resulted in surgery
# ---
# FLLINST
# Fall resulted in institutionalization
fall_columns = ['FLLDRVIS', 'FLLERVIS', 'FLLHOSP', 'FLLSURG','FLLINST' ]

#Â what are the range of values?
fall_unique_values = []
for x in fall_columns:
    fall_unique_values.append(pandas.unique(falls[x]).tolist())
    
print(fall_unique_values)

# Tag & count number of sig fall data entries

def sigfall(row):
    if row['FLLDRVIS'] == 1.0:
        return 1.0
    if row['FLLERVIS'] == 1.0:
        return 1.0
    if row['FLLHOSP'] == 1.0:
        return 1.0
    if row['FLLSURG'] == 1.0:
        return 1.0
    if row['FLLINST'] == 1.0:
        return 1.0
    else:
        return 0.0
    

falls['SIGFALL'] = falls.apply (lambda row: sigfall(row), axis=1)
falls['SIGFALL'].value_counts()
```

```python
updrs_sigfall_base['SIGFALL_12M'].value_counts()


# Re-doing to get NEXT sigfall, +12 month too vague 

def date_to_tuple(date):
    year = date[-4:]
    month = date[:2]
    return (month,year,date)

def date_sort(dates):
    tup_list = map(date_to_tuple,dates)
    return sorted(tup_list, key = lambda x: (x[1],x[0]))

def get_index(date,dates):
    output = (False,999)    
    for idx, val in enumerate(dates):
            if val[2] == date:
                output = (True,idx)
    return output

def next_sigfall(row):
    output = numpy.nan
    patno = row['PATNO']
    infodt = row['INFODT']    
    dates = get_infodts(patno)
    sorted_dates = date_sort(dates)
    i = get_index(infodt,sorted_dates)
    
    if i[0] == True:
        try:
            next_date_index = i[1] + 1
            next_date = sorted_dates[next_date_index]
            output = updrs_sigfall_base[(updrs_sigfall_base.INFODT == next_date ) & (updrs_sigfall_base.PATNO == patno)]['SIGFALL'].values[0]
        except:
            print()
    return output


# bum = date_sort(get_infodts(3326))
# print(bum)
# get_index('07/2014',bum)

        
updrs_future_sigfall = updrs_sigfall_base.copy()
updrs_future_sigfall['SIGFALL_NEXT'] = updrs_future_sigfall.apply (lambda row: next_sigfall(row), axis=1)
updrs_future_sigfall.to_csv('./working_data/updrs_future_sigfall.csv')
```
